```python
img=cv2.imread('0 (239).jpg',0)
print(img.shape)
```

```python
height,width=img.shape
plt.imshow(img,cmap='gray')
```

```python
#First let us flatten the image
imgFlatten = img.reshape(height*width,1)
print(imgFlatten.shape)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=8) #initially try starting from 3
model.fit(imgFlatten)
labels = model.labels_
print(labels.shape)
labels2D = labels.reshape(height,width)
print(img[200:210,200:210].T)
print("--------------------------------")
print(labels2D[200:210,200:210])
```

```python
plt.imshow(labels2D)
plt.colorbar()
plt.show()
```

```python
import joblib
joblib.dump(model,'brain_tumor_KMeans_model.sav')
```
